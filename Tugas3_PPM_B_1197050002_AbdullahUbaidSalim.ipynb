```python
# %%
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/Colab Notebooks/PPM/3/DatasetForCoffeeSales2.csv'
data = pd.read_csv(file_path)

# %%
display(data.head(10))
# %%
data.filter(["City", "Category", "Product", "Quantity"]).head()
# %%
data.sort_values("Quantity", axis=0, ascending=False, inplace=True, na_position="last")
display(data.head())
# %%
grp1 = data.groupby('City')
result = grp1['Final Sales'].aggregate('sum')
print(result)
```

```python
# %%
import pandas as pd

# %%
# reading the database


file_path = '/content/drive/My Drive/Colab Notebooks/PPM/3/students.csv'
data = pd.read_csv(file_path)

# %%
# printing the top 10 rows
display(data.head(10))

# %%
# Show the dataset info
data.info()

# %%
# Missing Value Checking
data.isna().sum()

# %%
# Filling Nan Value with mode
data.fillna({"lunch": data["lunch"].mode()[0]}, inplace=True)

#%%
# Filling Nan Value with mean
data.fillna({"reading score": data["reading score"].mean()}, inplace=True)

# %%
# Filling Nan Value with mode
data.fillna({"grade": data["grade"].median()}, inplace=True)

# %%
# Check the information of the dataset
data.info()

# %%
# printing the top 10 rows
display(data.head(10))
```
